```python
def get_cdf(x):
    cdf=[]
    d=1
    while d<=100:
        count = 0
        for i in x:
            for j in i :
                if j < (0.1*d):
                    count=count+1
        a = count/len(x)
        cdf.append(a)
        d=d+1
    return cdf
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.models import load_model
```

## Thử nghiệm với các model

### Model 1 

```python
LSTM_model = Sequential()

LSTM_model.add(LSTM(units = 60, return_sequences = True, input_shape = (X_train.shape[1], 1)))
LSTM_model.add(Dropout(0.1))

LSTM_model.add(LSTM(units = 60, return_sequences = True))
LSTM_model.add(Dropout(0.1))

LSTM_model.add(LSTM(units = 80))
LSTM_model.add(Dropout(0.1))

LSTM_model.add(Dense(units = 1))
```

```python
%%time
LSTM_model.compile(optimizer='adam', loss='mean_squared_error')
LSTM_model.fit(X_train,y_train, epochs=50, batch_size=32)
```

```python
y_pred_LSTM_model = LSTM_model.predict(X_test)
```

```python
y_pred_LSTM_model = scaler.inverse_transform(y_pred_LSTM_model)
```

```python
y_test.shape = (y_test.size//1,1)
y_test = scaler.inverse_transform(y_test)
```

```python
y_test
```

```python
# Visualising the results
plt.figure(figsize=(14,5))
plt.plot(y_test, color = 'red', label = 'Actual')
plt.plot(y_pred_LSTM_model, color = 'blue', label = 'Forecast')
plt.title('Compare Weather Forecast vs Actual')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
lose_in_model_1=abs(y_test-y_pred_LSTM_model)
```

```python
mean = np.mean(lose_in_model_1)
min = np.amin(lose_in_model_1) 
max = np.amax(lose_in_model_1) 
range = np.ptp(lose_in_model_1) 
varience = np.var(lose_in_model_1) 
sd = np.std(lose_in_model_1) 
print("Measures of Dispersion")
print("Mean=", mean)
print("Minimum =", min) 
print("Maximum =", max) 
print("Range =", range) 
print("Varience =", varience) 
print("Standard Deviation =", sd) 
```
